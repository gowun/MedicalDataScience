```
def average_cols(df):
  new_df = pd.DataFrame()
  cols = sorted(set(map(lambda x: str(x), df.columns)))
  col_cnts = pd.value_counts(df.columns)
  for c in cols:
    if c != 'nan':
      if col_cnts[c] > 1: 
        new_df[c] = df[c].mean(1)
      else:
        new_df[c] = df[c]
  return new_df
```

```
def plot_boxes(df_list, cols, label_list):
  print('Gene Expressions Taken')
  print(cols)
  not_in_all = []
  for c in cols:
    arr = list(df_list[0][c])
    labels = [label_list[0]] * len(arr)
    try:
      for i in range(1, len(df_list)):
        arr += list(df_list[i][c])
        labels += ([label_list[i]] * len(df_list[i]))
      fe.plot_box(arr, 'x', labels, c)
    except KeyError:
      not_in_all.append(c)
  print('\n\nGene Expressions not in Two Sides')
  print(not_in_all)
```
